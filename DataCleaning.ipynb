```python
acidentes = pd.read_excel('traffic_accidents.xlsx')
```

```python
acidentes.head(3)
```

```python
acidentes[['light_condition', 'road_lighting']] = acidentes['lighting_condition'].str.split(',', expand=True)
```

```python
acidentes.drop(columns = ['lighting_condition'], inplace=True)
```

```python
acidentes[['injury', 'attitude']] = acidentes['crash_type'].str.split(' / ', expand=True)
```

```python
acidentes.drop(columns = ['crash_type'], inplace=True)
```

```python
acidentes['crash_date'] = pd.to_datetime(acidentes['crash_date'])
```

```python
acidentes['date'] = acidentes['crash_date'].dt.date
```

```python
acidentes['time'] = acidentes['crash_date'].dt.time
```

```python
acidentes.drop(columns = ['crash_date'], inplace=True)
```

```python
acidentes['damage'] = acidentes['damage'].str.replace('$', '', regex = False)
acidentes['damage'] = acidentes['damage'].str.replace(',', '', regex = False)
```

```python
def clean_cost(value):
  #caso com -
  if '-' in value:
    parts = value.split('-')
    low = float(parts[0]) #valor minimo
    high = float(parts[1]) #valor máximo
    return (low + high) / 2
  elif 'OVER' in value:
    return float(value.split(' ')[-1])
  elif 'OR LESS' in value:
    return float(value.split(' ')[0])
```

```python
acidentes['damage'] = acidentes['damage'].apply(clean_cost)
```

```python
days_week = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}

name_month = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
```

```python
acidentes['crash_day_of_week'] = acidentes['crash_day_of_week'].map(days_week)
```

```python
acidentes['crash_month'] = acidentes['crash_month'].map(name_month)
```
